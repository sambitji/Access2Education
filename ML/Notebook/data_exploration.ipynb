# 📊 Notebook 1: Data Exploration
### Edu-Platform - Student Aptitude Data Analysis

**Goal:** Samajhna ki hamare aptitude test ka data kaisa dikhta hai, koi missing values toh nahi, aur features kaisi distribute hain.

**Steps:**
1. Synthetic data generate karna (jab tak real data nahi aata)
2. Basic statistics dekhna
3. Distributions visualize karna
4. Correlation matrix banana
5. Outliers identify karna
6. Clean data CSV mein save karna

```python
# ✅ Step 0: Libraries Import
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
import warnings
warnings.filterwarnings('ignore')

# Plot style set karo
plt.style.use('seaborn-v0_8-whitegrid')
sns.set_palette('Set2')

print('✅ Libraries successfully import ho gayi!')
```

## 📌 Step 1: Synthetic Data Generate Karna

Humara aptitude test 5 cheezein measure karega:
- **Logical Reasoning** → Problem solving ability
- **Verbal Ability** → Reading/understanding
- **Numerical Ability** → Math/calculations
- **Memory Retention** → Kitna yaad rakhta hai
- **Attention Span** → Focus level

Sabka score 0-100 ke beech hoga.

```python
# ✅ Step 1: Synthetic Student Data Generate Karo
np.random.seed(42)  # Reproducibility ke liye
N = 500  # 500 students ka data

# 4 alag-alag types ke students create karo (real mein ye clusters honge)
# Type 1: Visual Learners - High verbal & memory, low numerical
type1 = pd.DataFrame({
    'student_id': [f'STU_{i:04d}' for i in range(1, 126)],
    'logical':    np.random.normal(55, 10, 125).clip(0, 100),
    'verbal':     np.random.normal(80, 8,  125).clip(0, 100),
    'numerical':  np.random.normal(45, 12, 125).clip(0, 100),
    'memory':     np.random.normal(78, 9,  125).clip(0, 100),
    'attention':  np.random.normal(70, 10, 125).clip(0, 100),
    'true_type':  'visual_learner'
})

# Type 2: Conceptual Thinkers - High logical, average baaki sab
type2 = pd.DataFrame({
    'student_id': [f'STU_{i:04d}' for i in range(126, 251)],
    'logical':    np.random.normal(85, 7,  125).clip(0, 100),
    'verbal':     np.random.normal(65, 10, 125).clip(0, 100),
    'numerical':  np.random.normal(70, 9,  125).clip(0, 100),
    'memory':     np.random.normal(55, 11, 125).clip(0, 100),
    'attention':  np.random.normal(60, 10, 125).clip(0, 100),
    'true_type':  'conceptual_thinker'
})

# Type 3: Practice-Based Learners - High numerical & attention
type3 = pd.DataFrame({
    'student_id': [f'STU_{i:04d}' for i in range(251, 376)],
    'logical':    np.random.normal(65, 9,  125).clip(0, 100),
    'verbal':     np.random.normal(50, 11, 125).clip(0, 100),
    'numerical':  np.random.normal(85, 7,  125).clip(0, 100),
    'memory':     np.random.normal(60, 10, 125).clip(0, 100),
    'attention':  np.random.normal(80, 8,  125).clip(0, 100),
    'true_type':  'practice_based'
})

# Type 4: Step-by-Step Learners - Balanced lekin memory high
type4 = pd.DataFrame({
    'student_id': [f'STU_{i:04d}' for i in range(376, 501)],
    'logical':    np.random.normal(60, 10, 125).clip(0, 100),
    'verbal':     np.random.normal(62, 10, 125).clip(0, 100),
    'numerical':  np.random.normal(58, 9,  125).clip(0, 100),
    'memory':     np.random.normal(85, 7,  125).clip(0, 100),
    'attention':  np.random.normal(75, 9,  125).clip(0, 100),
    'true_type':  'step_by_step'
})

# Sabko combine karo
df = pd.concat([type1, type2, type3, type4], ignore_index=True)
df = df.sample(frac=1, random_state=42).reset_index(drop=True)  # Shuffle karo

print(f'✅ Dataset shape: {df.shape}')
print(f'Total students: {len(df)}')
print(f'\nStudent types distribution:')
print(df['true_type'].value_counts())
```

```python
# ✅ Step 2: Basic Dataset Information
print('='*60)
print('📋 DATASET KI BASIC INFORMATION')
print('='*60)
print(df.head(10))
```

```python
# Data types aur missing values check karo
print('\n📊 Data Types:')
print(df.dtypes)

print('\n❓ Missing Values (agar 0 hai toh sab theek):')
print(df.isnull().sum())

print('\n📈 Basic Statistics:')
print(df.describe().round(2))
```

## 📌 Step 3: Feature Distributions Visualize Karo

```python
# ✅ Step 3a: Har feature ki distribution plot karo
features = ['logical', 'verbal', 'numerical', 'memory', 'attention']
feature_names = ['Logical Reasoning', 'Verbal Ability', 'Numerical Ability', 'Memory Retention', 'Attention Span']

fig, axes = plt.subplots(2, 3, figsize=(15, 10))
axes = axes.flatten()

colors = ['#3498db', '#e74c3c', '#2ecc71', '#f39c12', '#9b59b6']

for i, (feat, name, color) in enumerate(zip(features, feature_names, colors)):
    axes[i].hist(df[feat], bins=30, color=color, alpha=0.7, edgecolor='white')
    axes[i].axvline(df[feat].mean(), color='black', linestyle='--', linewidth=2, label=f'Mean: {df[feat].mean():.1f}')
    axes[i].axvline(df[feat].median(), color='red', linestyle='-', linewidth=2, label=f'Median: {df[feat].median():.1f}')
    axes[i].set_title(f'{name}', fontsize=13, fontweight='bold')
    axes[i].set_xlabel('Score (0-100)')
    axes[i].set_ylabel('Number of Students')
    axes[i].legend(fontsize=9)

# Last subplot hide karo
axes[-1].set_visible(False)

plt.suptitle('📊 Aptitude Test Score Distributions - 500 Students', fontsize=16, fontweight='bold', y=1.02)
plt.tight_layout()
plt.savefig('../plots/01_score_distributions.png', dpi=150, bbox_inches='tight')
plt.show()
print('✅ Distribution plots save ho gayi!')
```

```python
# ✅ Step 3b: Student Types ke hisaab se box plots
import os
os.makedirs('../plots', exist_ok=True)

fig, axes = plt.subplots(1, 5, figsize=(20, 6))

type_colors = {
    'visual_learner': '#3498db',
    'conceptual_thinker': '#e74c3c',
    'practice_based': '#2ecc71',
    'step_by_step': '#f39c12'
}

for i, (feat, name) in enumerate(zip(features, feature_names)):
    data_by_type = [df[df['true_type'] == t][feat].values for t in type_colors.keys()]
    bp = axes[i].boxplot(data_by_type, patch_artist=True, notch=True)
    
    for patch, color in zip(bp['boxes'], type_colors.values()):
        patch.set_facecolor(color)
        patch.set_alpha(0.7)
    
    axes[i].set_title(name, fontsize=11, fontweight='bold')
    axes[i].set_xticklabels(['Visual', 'Conceptual', 'Practice', 'Step-by-Step'], rotation=30, fontsize=8)
    axes[i].set_ylabel('Score')

plt.suptitle('📦 Score Distribution by Learning Type', fontsize=14, fontweight='bold')
plt.tight_layout()
plt.savefig('../plots/01_boxplots_by_type.png', dpi=150, bbox_inches='tight')
plt.show()
```

## 📌 Step 4: Correlation Analysis

```python
# ✅ Step 4: Correlation Heatmap
fig, axes = plt.subplots(1, 2, figsize=(16, 6))

# Overall correlation
corr_matrix = df[features].corr()
mask = np.triu(np.ones_like(corr_matrix, dtype=bool))

sns.heatmap(
    corr_matrix, 
    mask=mask,
    annot=True, 
    fmt='.2f',
    cmap='RdYlGn',
    center=0,
    square=True,
    ax=axes[0],
    cbar_kws={"shrink": 0.8},
    linewidths=0.5
)
axes[0].set_title('🔗 Feature Correlation Matrix', fontsize=13, fontweight='bold')

# Average scores per student type
avg_by_type = df.groupby('true_type')[features].mean()
avg_by_type.columns = ['Logical', 'Verbal', 'Numerical', 'Memory', 'Attention']

sns.heatmap(
    avg_by_type,
    annot=True,
    fmt='.1f',
    cmap='YlOrRd',
    ax=axes[1],
    cbar_kws={"shrink": 0.8},
    linewidths=0.5
)
axes[1].set_title('🎯 Avg Score by Learning Style', fontsize=13, fontweight='bold')

plt.tight_layout()
plt.savefig('../plots/01_correlation_heatmaps.png', dpi=150, bbox_inches='tight')
plt.show()

print('\n📊 Average Scores by Learning Type:')
print(avg_by_type.round(2))
```

## 📌 Step 5: Outlier Detection

```python
# ✅ Step 5: Outlier Detection using Z-Score
print('🔍 Outlier Detection (Z-Score method, threshold = 3)\n')

outlier_counts = {}
for feat in features:
    z_scores = np.abs(stats.zscore(df[feat]))
    outliers = (z_scores > 3).sum()
    outlier_counts[feat] = outliers
    print(f'{feat.capitalize():20s}: {outliers} outliers')

# Outliers ko remove karo (z-score > 3)
z_scores_all = np.abs(stats.zscore(df[features]))
mask_no_outlier = (z_scores_all < 3).all(axis=1)
df_clean = df[mask_no_outlier].copy()

print(f'\n📊 Original data size: {len(df)}')
print(f'📊 Clean data size:    {len(df_clean)}')
print(f'🗑️  Removed outliers:   {len(df) - len(df_clean)}')
```

```python
# ✅ Step 6: Pairplot - Sabke beech relationship
# Sirf 150 samples lo warna slow hoga
sample_df = df_clean.sample(150, random_state=42)

type_palette = {
    'visual_learner': '#3498db',
    'conceptual_thinker': '#e74c3c', 
    'practice_based': '#2ecc71',
    'step_by_step': '#f39c12'
}

pair_plot = sns.pairplot(
    sample_df[features + ['true_type']], 
    hue='true_type',
    palette=type_palette,
    plot_kws={'alpha': 0.6, 's': 30},
    diag_kind='kde'
)
pair_plot.fig.suptitle('🔍 Pairplot: Feature Relationships by Learning Type', y=1.02, fontsize=14, fontweight='bold')
plt.savefig('../plots/01_pairplot.png', dpi=120, bbox_inches='tight')
plt.show()
print('✅ Pairplot save ho gayi!')
```

## 📌 Step 7: Clean Data Save Karo

```python
# ✅ Step 7: Clean data CSV mein save karo
import os
os.makedirs('../../data', exist_ok=True)

df_clean.to_csv('../../data/sample_aptitude_data.csv', index=False)
print(f'✅ Clean data save ho gaya: data/sample_aptitude_data.csv')
print(f'   Total records: {len(df_clean)}')
print(f'   Columns: {list(df_clean.columns)}')

# Final summary
print('\n' + '='*60)
print('📝 EXPLORATION SUMMARY')
print('='*60)
print(f'✅ Total students analyzed: {len(df_clean)}')
print(f'✅ Features used: {features}')
print(f'✅ No missing values found')
print(f'✅ Outliers removed: {len(df) - len(df_clean)}')
print(f'✅ 4 distinct learning patterns clearly visible in pairplot')
print(f'✅ Data ready for clustering!')
```
